```python
import pandas
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#coords = pandas.read_table('/home/users/jvogel/Science/hipp_tau_clus/string_network_coordinates_all_EntrezIDs.txt') 
#ints = pandas.read_csv('/home/users/jvogel/Science/hipp_tau_clus/string_interactions_all_EntrezIDs.csv', sep='\t')
coords = pandas.read_table('/Users/jakevogel/Downloads/string_network_coordinates.txt')
ints = pandas.read_csv('/Users/jakevogel/Downloads/string_interactions.tsv', sep='\t')
```

```python
df = pandas.DataFrame(np.zeros((len(coords['#node']),len(coords['#node']))),
                      index = coords['#node'], columns = coords['#node'])  
```

```python
for i,row in ints.iterrows():                                           
    df.loc[row['#node1'],row['node2']] = row['combined_score']
df.head()
```

```python
plt.close()
sns.clustermap(df)
plt.show()
```

```python
tst = nx.from_numpy_matrix(df.values)
```

```python
#tst = nx.from_pandas_dataframe(df)
```

```python
hubbies = nx.hub_matrix(tst)
```

```python
hubdf = pandas.DataFrame(index = df.index, columns = ['hubness'])
for i in range(len(df.index)):
    hubdf.loc[df.index[i], 'hubness'] = hubbies[i,i]
```

```python
hubdf.sort_values('hubness', ascending=False)
```
