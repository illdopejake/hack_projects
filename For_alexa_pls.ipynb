```python
import pandas
import os
import nibabel as ni
import numpy as np
import difflib
```

```python
labs = pandas.read_csv('/Users/jakevogel/git/pycourse_vuMC/stuff/dst_labels.csv', header = None)
labs.newlabs = [x.lower().strip() for x in labs[labs.columns[-1]]]
```

```python
labs['newlabs'] = [x.lower().replace(' ','') for x in labs[labs.columns[-1]]]
```

```python
labs.index = labs.newlabs
```

```python
alexa = pandas.ExcelFile('/Users/jakevogel/Downloads/amyloid_PLS_loadings.xlsx').parse('Sheet1')
alexa.head()
```

```python
for i,row in alexa.iterrows():
    lab_str = row['FreeSurfer regions']
    lab = lab_str.split('-')[-1]
    match = difflib.get_close_matches(lab,labs.newlabs)
    if len(match)>0:
        match = match[0]
    else:
        continue
    indices = labs.ix[match,labs.columns[0]].values
    alexa.ix[i,'lab1'] = indices[0]
    alexa.ix[i,'lab2'] = indices[-1]
```

```python
alexa
```

```python
jnk = ni.load('/Users/jakevogel/git/pycourse_vuMC/stuff/dkt_atlas_1mm.nii.gz')
aff = jnk.affine
atlas = jnk.get_data()
```

```python
for i,row in alexa.iterrows():
    if pandas.notnull(row['lab1']):
        atlas[atlas == row['lab1']] = row['loadings_psych_personality']
        atlas[atlas == row['lab2']] = row['loadings_psych_personality']

# we need to get rid of the other values that were not used in the spreadsheet
# since all the "good" values are negative numbers, we can use a hack:
# we'll just set everything above 0 to 0
atlas[atlas > 0] = 0
```

```python
nimg = ni.Nifti1Image(atlas,aff)
nimg.to_filename('/Users/jakevogel/Downloads/alexa_pls_projection')
```
